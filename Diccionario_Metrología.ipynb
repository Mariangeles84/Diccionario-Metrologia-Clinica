ESTADISTICA DESCRIPTIVA CON STATISTICS

```python
import statistics as stats
```

```python
media = stats.mean(measurement_col)
print("Media: ", media)

desviacion = stats.stdev(measurement_col)
print("Desviación estándar: ", desviacion)

q1 = stats.quantiles(measurement_col)[0]
print("Primer cuartil: ", q1)

q2 = stats.quantiles(measurement_col)[1]
print("Segundo cuartil: ", q2)

q3 = stats.quantiles(measurement_col)[2]
print("Tercer cuartil: ", q3)

```

```python
media = stats.mean(length_col)
print("Media: ", media)

desviacion = stats.stdev(length_col)
print("Desviación estándar: ", desviacion)

q1 = stats.quantiles(length_col)[0]
print("Primer cuartil: ", q1)

q2 = stats.quantiles(length_col)[1]
print("Segundo cuartil: ", q2)

q3 = stats.quantiles(length_col)[2]
print("Tercer cuartil: ", q3)
```
